```
import matplotlib.pyplot as plt # for plotting
import matplotlib.patches as mpatch
```

```
def popsize_vs_num_iters(population_size, num_iterations, time, y_log_scale=True, file_name="pop_plot.png"):
  fig, ax1 = plt.subplots()

  plt.title('Population Size vs Number of Iterations and Computation Time')
  plt.grid()

  ax1.set_xlabel('Population Size')
  ax1.set_ylabel('Mean Number of Iterations')
  ax1.plot(population_size, num_iterations, color = 'r')
  ax1.scatter(population_size, num_iterations, color = 'r')
  ax2 = ax1.twinx()

  ax2.set_ylabel('Average Computation Time (sec)')
  ax2.plot(population_size, time, color = 'b')
  ax2.scatter(population_size, time, color = 'b')
  
  fig.tight_layout()

  line1 = mpatch.Patch(color='r')
  line2 = mpatch.Patch(color='b')
  plt.legend([line1, line2],["Mean Number of Iterations", "Average Computation Time"])

  if y_log_scale:
    ax1.set_yscale('log')
    ax2.set_yscale('log')
  
  if file_name:
    plt.savefig(file_name)
  plt.show()
```

```
pop = [5, 10, 15, 20, 25]
num_iter = [528.0546, 30.94171, 24.91111, 20.90152, 20.46053]
mean_time = [2.533055, 0.252136, 0.286848, 0.298739, 0.373299]

popsize_vs_num_iters(pop, num_iter, mean_time, y_log_scale=False)
```

```
def mutation_prob_vs_iterations(num_iterations, mutation_probs,
                                file_name="mut_plot.png"):
  fig, ax = plt.subplots()
  plt.plot(mutation_probs, num_iterations)
  plt.scatter(mutation_probs, num_iterations)
  plt.xlabel('Mutation Probability')
  plt.ylabel('Average Number of Iterations')
  plt.title('Effect of Mutation Probability on Average Number of Iterations')
  plt.grid()

  if file_name:
    plt.savefig(file_name)
  plt.show()
```

```
mut = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3]
num_iter = [37.71831, 31.20224, 35.49396, 36.59128, 33.85152, 35.55625]

mutation_prob_vs_iterations(num_iter, mut)
```
