```
import matplotlib.pyplot as plt # for plotting
import numpy as np # for synthesizing fake data during development
```

```
def attack_success_rate_plot(attack_success_rate, dataset_names,
                             file_name="plot1.png", ylim=(0,1)):
  """
    Attack success rate bar plots vs dataset names.
    Adapts to arbitarility many datasets.
    If the dataset names overlap, include the \\n character in your string
  """
  # fig, ax = plt.subplots()
  assert len(attack_success_rate) == len(dataset_names), "Incorrect lengths"
  plt.ylim(ylim)
  x = np.arange(len(dataset_names))
  plt.bar(x, attack_success_rate)
  plt.xticks(x, dataset_names)
  plt.grid()
  plt.title('Adversarial Attack Success Rate vs. Dataset')
  if file_name:
    plt.savefig(file_name)
  plt.show()
```

```
def popsize_vs_num_iters(population_size, num_iterations,y_log_scale=True, file_name="plot2.png"):
  fig, ax = plt.subplots()
  plt.plot(population_size, num_iterations)
  plt.scatter(population_size, num_iterations)
  plt.xlabel('Population Size')
  plt.ylabel('Minimum Number of Iterations')
  plt.title('Minimum Number of Iterations vs Population Size in Genetic Attack')
  plt.grid()

  if y_log_scale:
    ax.set_yscale('log')
  
  if file_name:
    plt.savefig(file_name)
  plt.show()
```

```
def mutation_prob_vs_iterations(num_iterations, mutation_probs,
                                file_name="plot3.png"):
  fig, ax = plt.subplots()
  plt.plot(mutation_probs, num_iterations)
  plt.scatter(mutation_probs, num_iterations)
  plt.xlabel('Mutation Probability')
  plt.ylabel('Minimum Number of Iterations')
  plt.title('Minimum Number of Iterations vs Population Size in Genetic Attack')
  plt.grid()
  # ax.set_xscale('log')
  if file_name:
    plt.savefig(file_name)
  plt.show()
```

```
def main():
  np.random.seed(69)
  # Plot 1: Adversarial attack rate vs. models
  attack_success_rate = np.random.rand(4)
  attack_success_rate
  model_names = ["MNIST", "Fashion-MNIST", "MNIST\n (Adv. Trained)", "Fashion-MNIST\n (Adv. Trained)"]
  attack_success_rate_plot(attack_success_rate, model_names)

  # Plot 2: Population size vs. # of iterations (Fig. 5 from paper)
  population_size = [10, 20, 50, 100]
  num_iterations =  [1e3, 2e3, 1e4, 5e5]
  popsize_vs_num_iters(population_size, num_iterations)

  # Plot 3: mutation rate vs min number of iterations
  num_iterations = [6e5, 4e5, 1e5, 5e5, 7e5]
  mutation_probs = [0.01, 0.1, 0.25, 0.75, 1.00]
  mutation_prob_vs_iterations(num_iterations, mutation_probs)

if __name__ == "__main__":
  main()
```
